```
text = ""

file_path = "books-data/kafka/metamorphosis.txt"

with open(file_path, 'r', encoding="utf-8-sig") as file:
    text = file.read()
print ('Length of text: {} characters'.format(len(text)))
```

```
# Preprocess data
punctuation_translator = str.maketrans('–—”„…«»‘’“°ſ†•✠', '               ', string.punctuation)
digits_translator = str.maketrans('', '', string.digits)
special_characters_translator = str.maketrans('ąćęłńóśźżäöüæèêéôâáà£çëîñòùúûāœï', 'acelnoszzaoueeeeoaaaeceinouuuaei', '')

# remove redundant characters and translate special characters
text = text.lower().translate(punctuation_translator).translate(digits_translator).translate(special_characters_translator)

# remove "chapter <number>" strings 
text = re.sub(r"\nchapter\s(.*)\n", "", text)

#remove extra spaces and new lines
text = ' '.join(text.split())

vocab = sorted(list(set(text)))
vocab_size = len(vocab)
print ('{} unique characters:'.format(len(vocab)))

words = text.split()
print('Total words:', len(words))
unique_words=sorted(set(words))
print('Unique words:', len(unique_words))
```

```
def make_pairs_n_1(words):
  for i in range(len(words) - 1):
    yield (words[i], words[i + 1])

def make_pairs_n_2(words):
  for i in range(len(words) - 2):
    yield (words[i] + " " + words[i+1], words[i + 2])

pairs = make_pairs_n_2(words)

word_dict = {}
for word_1, word_2 in pairs:
  if word_1 in word_dict.keys():
    word_dict[word_1].append(word_2)
  else:
    word_dict[word_1] = [word_2]
#######################################################################
from collections import Counter

for key in word_dict:
  cnt = Counter(word_dict[key])
  word_dict[key] = cnt.most_common(1)[0][0]
```

```
next_input = "years that"
chain = [next_input]
 
n_words = 20
for i in range(n_words):
  # chain.append(np.random.choice(word_dict[chain[-1]]))
  # chain.append(word_dict[chain[-1]])

  second_word = next_input.split()[1]
  prediction = word_dict[next_input]
  chain.append(prediction)
  next_input = second_word + " " + prediction

print(' '.join(chain))
```
